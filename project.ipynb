```python
np.random.seed(42)
data['Height']=np.random.randint(150,181,size=data.shape[0])
data
```

```python
# 1.How many are there in each Team and the percentage splitting with respect to the total employees.
```

```python
team_count=data['Team'].value_counts()
team_count
```

```python
team_percentages=(team_count/len(data))*100
team_percentages
```

```python
# 2.Segregate the employees w.r.t different positions.
```

```python
position_count=data['Position'].value_counts()
position_count
```

```python
# 3.Find from which age group most of the employees belong to.
```

```python
age_groups = pd.cut(data['Age'],
       bins=[20, 30, 40, 50, 60, 70], 
     labels=['20-29', '30-39', '40-49', '50-59', '60-69'])
common_age_group = age_groups.value_counts().idxmax()
common_age_group
```

```python
# 4.Find out under which team and position, spending in terms of salary is high.
```

```python
salary = data.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
salary
```

```python
# 5.Find if there is any correlation between age and salary , represent it visually
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Age', y='Salary', data=data)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
